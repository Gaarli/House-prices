```python
import pandas as pd

df = pd.read_csv('kc_house_data.csv')
```

```python
# ID como string: identificador, não valor numérico (variável categórica nominal)
df['id'] = df['id'].astype(str)

# Date como datetime: facilita análises temporais
df['date'] = pd.to_datetime(df['date'])

# Zipcode como string: evita perda de zeros à esquerda (variável categórica nominal)
df['zipcode'] = df['zipcode'].astype(str)

# Waterfront como bool: variável binária (True/False)
df['waterfront'] = df['waterfront'].astype(bool)

# Condition como category: poucos valores repetidos, economiza memória
df['condition'] = df['condition'].astype('category')

# Grade como category: variável categórica com valores discretos
df['grade'] = df['grade'].astype('category')
```
